```python
import numpy as np

# Target function: y = 2*x1 + 3*x2 - x3 + 5
def target_function(x):
    return 2*x[0] + 3*x[1] - x[2] + 5

def generate_dataset(n_samples=10):
    X = np.random.rand(n_samples, 3)
    y = np.array([target_function(x) for x in X])
    return X, y

def train_linear_perceptron(X, y, lr=0.01, epochs=100):
    weights = np.zeros(X.shape[1])
    bias = 0
    n = len(y)
    for epoch in range(epochs):
        predictions = X.dot(weights) + bias
        errors = y - predictions
        mse = np.mean(errors ** 2)
        weights += lr * X.T.dot(errors) / n
        bias += lr * errors.mean()
        print(f"Epoch {epoch+1}: MSE = {mse:.4f}")
    return weights, bias

def main():
    np.random.seed(42)
    X, y = generate_dataset()
    weights, bias = train_linear_perceptron(X, y)
    print("\nFinal Weights:", weights)
    print("Final Bias:", bias)

if __name__ == "__main__":
    main()
```
